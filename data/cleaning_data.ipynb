```python
import pandas as pd
import numpy as np
```

```python
df_temp = pd.read_csv('./temp_country_group.csv')
df_temp = df_temp.set_index(['ISO Code'])

df_temp_1950 = df_temp[[x >= 1950 for x in df_temp.dt]].reset_index()

df_temp_1950_pivot = df_temp_1950.pivot(index='ISO Code', columns='dt', values='AverageTemperature')

df_temp_1950 = df_temp_1950.set_index(['ISO Code', 'dt'])
df_temp_1950.head()
```

```python
df_delta_temp = df_temp_1950_pivot.copy()

df_1950 = df_delta_temp[1950].copy()

for col in df_delta_temp.columns:
    df_delta_temp[col] = np.around(df_delta_temp[col] - df_1950, decimals=5)

df_delta_temp = df_delta_temp.reset_index()

#df_delta_temp.to_csv('delta_temp.csv', encoding='utf-8', index=False)
#df_delta_temp = pd.read_csv('delta_temp.csv')

df_melt = pd.melt(df_delta_temp, id_vars='ISO Code', value_vars=list(df_delta_temp.columns[1:]), var_name='dt', value_name='delta')

df_melt = df_melt.set_index(['ISO Code', 'dt'])
df_melt = df_melt.sort_index()
df_melt.head()
```

```python
df_merged = df_melt.join(df_temp_1950)
df_merged.head()
```

```python
df_pop = pd.read_excel('population_country.xls')
df_pop =df_pop.dropna(how='all', axis=1)
df_pop = df_pop.drop(['Country Name', 'Indicator Name', 'Indicator Code'], axis=1)

#df_pop.to_csv('population_country.csv', index=False, encoding='utf-8')
#df_pop = pd.read_csv('population_country.csv')

df_melt2 = pd.melt(df_pop, id_vars='Country Code', value_vars=list(df_pop.columns[1:]), var_name='dt', value_name='population')
df_melt2.dt = df_melt2.dt.astype(int)
df_melt2 = df_melt2.rename(columns={'Country Code':'ISO Code'})
df_melt2 = df_melt2.set_index(['ISO Code', 'dt'])
df_melt2 = df_melt2.sort_index()
df_melt2.head()
```

```python
df_merged2 = df_melt2.join(df_merged)
df_merged2.head()
```

```python
df_co2 = pd.read_excel('CO2_country.xls')
df_co2 = df_co2.dropna(how='all', axis=1)
df_co2 = df_co2.drop(['Country Name', 'Indicator Name', 'Indicator Code'], axis=1)

#df_co2.to_csv('CO2_country.csv', index=False, encoding='utf-8')
#df_co2 = pd.read_csv('CO2_country.csv')

df_melt3 = pd.melt(df_co2, id_vars='Country Code', value_vars=list(df_pop.columns[1:]), var_name='dt', value_name='CO2')
df_melt3.dt = df_melt3.dt.astype(int)
df_melt3 = df_melt3.rename(columns={'Country Code':'ISO Code'})
df_melt3 = df_melt3.set_index(['ISO Code', 'dt'])
df_melt3 = df_melt3.sort_index()
df_melt3.head()
```

```python
df_merged3 = df_melt3.join(df_merged2)
df_merged3 = df_merged3.reset_index()

df_merged3.to_csv('full_data.csv', index=False, encoding='utf-8')
df_full_data = pd.read_csv('full_data.csv')

df_full_data
```
